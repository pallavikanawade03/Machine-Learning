```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
temperature = pd.read_csv("/content/temperatures.csv")
temperature.head()
```

```python
temperature.plot(x = 'YEAR', y = 'ANNUAL', kind = 'scatter')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
x = temperature["YEAR"].values.reshape(-1, 1)
y = temperature["ANNUAL"].values.reshape(-1, 1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
LinearRegression()
```

```python
y_pred = regressor.predict(x_test)
```

```python
print(regressor.intercept_)
print(regressor.coef_)
```

```python
plt.scatter(x_test, y_test, color = "green")
plt.plot(x_test, y_pred, color = "orange")
plt.show()
```

```python
print("Mean Absolute Error: ", metrics.mean_absolute_error(y_test, y_pred))
print("Mean Squared Error: ", metrics.mean_squared_error(y_test, y_pred))
print("R2 Score: ", metrics.r2_score(y_test, y_pred))
```

```python
import numpy as np
import pandas as pd
```

```python
temperature = pd.read_csv("/content/temperatures.csv")
temperature.head()
```

```python
# Remove Duplicates
temperature = temperature.drop_duplicates()
print(temperature)
```

```python
# Checking blank value

temperature.isnull()
```

```python
# Checking null values category wise

temperature.isnull().sum()
```

```python
# Dropping categorical data rows with missing values

temperature.dropna(how = 'any', inplace = True)
```

```python
# Checking null values category wise

temperature.isnull().sum()
```

```python
# Splitting Dataset

x = temperature[["YEAR", "JAN", "FEB"]].values
y = temperature["JAN-FEB"].values
print(x)
```

```python
print(y)
```

```python
from sklearn.impute import SimpleImputer
print(x[:, 1:3])
```

```python
imputer = SimpleImputer(missing_values = np.nan, strategy='mean')
imputer.fit(x[:, 1:3])
x[:, 1:3] = imputer.transform(x[:, 1:3])
print(x[:, 1:3])
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

ct = ColumnTransformer(transformers= [('encoder', OneHotEncoder(), [0])], remainder = 'passthrough')
x = np.array(ct.fit_transform(x))
print(x)
```
