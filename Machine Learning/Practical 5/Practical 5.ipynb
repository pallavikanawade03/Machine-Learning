```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
crashes = pd.read_csv("/content/car_crashes.csv")
crashes.head()
```

```python
plt.plot(crashes.speeding, crashes.total)
plt.xlabel("Speeding")
plt.ylabel("Total")
plt.title("Plot")
plt.show()
```

```python
plt.scatter(crashes.alcohol, crashes.total)
plt.xlabel("Alcohol")
plt.ylabel("Total")
plt.title("Scatter")
plt.show()
```

```python
plt.bar(crashes.alcohol, crashes.total)
plt.xlabel("Alcohol")
plt.ylabel("Total")
plt.title("Bar")
plt.show()
```

```python
plt.stem(crashes.not_distracted, crashes.total)
plt.xlabel("Not Disracted")
plt.ylabel("Total")
plt.title("Stem")
plt.show()
```

```python
plt.fill_between(crashes.total, crashes.alcohol, crashes.not_distracted)
plt.xlabel("Total")
plt.title("Fill Between")
plt.show()
```

```python
plt.stairs(crashes.total, linewidth=2.5)
plt.title("Stack plot")
plt.show()
```

```python
plt.hist(crashes.total, bins=8, linewidth=0.5, edgecolor="white")
plt.title("Hist")
plt.show()
```

```python
crashes.boxplot(figsize = (10, 10), column =['total'])
```
